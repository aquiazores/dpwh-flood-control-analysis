```python
print(df['ApprovedBudgetForContract'].head(10))
print("---")
print(df['ContractCost'].head(10))
```

```python
df['ApprovedBudgetForContract'] = pd.to_numeric(df['ApprovedBudgetForContract'])
df['ContractCost'] = pd.to_numeric(df['ContractCost'])

print(df['ApprovedBudgetForContract'].dtype)
print(df['ContractCost'].dtype)
```

```python
df['ActualCompletionDate'] = pd.to_datetime(df['ActualCompletionDate'])
df['StartDate'] = pd.to_datetime(df['StartDate'])

print(df['ActualCompletionDate'].dtype)
print(df['StartDate'].dtype)
```

```python
df['Municipality'] = df['Municipality'].fillna('Unknown')

print(df['Municipality'].isnull().sum())
```

```python
df['CostVariance'] = df['ContractCost'] - df['ApprovedBudgetForContract']

df['CostVariancePct'] = (df['CostVariance'] / df['ApprovedBudgetForContract']) * 100

df['ProjectDurationDays'] = (df['ActualCompletionDate'] - df['StartDate']).dt.days

print(df[['ApprovedBudgetForContract', 'ContractCost','CostVariance', 'CostVariancePct', 'ProjectDurationDays']].head(10))
```

```python
print("Total rows and columns:", df.shape)
print("\nColumn data types:")
print(df.dtypes)
print("\nRemaining nulls:")
print(df.isnull().sum())
```

```python

problem_rows = df[df['ApprovedBudgetForContract'].isnull()]

print("Number of problem rows:", len(problem_rows))
print(problem_rows[['ProjectName', 'ApprovedBudgetForContract', 'ContractCost']].head(20))
```

```python

df_excluded = df[df['ApprovedBudgetForContract'].isnull()]

print("Excluded rows:", len(df_excluded))
```

```python
df_clean = df.dropna(subset=['ApprovedBudgetForContract', 'ContractCost']).copy()

print("Rows available for analysis:", len(df_clean))
```

```python
print("Remaining nulls:")
print(df_clean.isnull().sum())
print("--")
print("Final row count:", len(df_clean))
```

```python
# Recalculate on clean data since we dropped rows
df_clean['CostVariance'] = df_clean['ContractCost'] - df_clean['ApprovedBudgetForContract']
df_clean['CostVariancePct'] = (df_clean['CostVariance'] / df_clean['ApprovedBudgetForContract']) * 100
df_clean['ProjectDurationDays'] = (df_clean['ActualCompletionDate'] - df_clean['StartDate']).dt.days

print(df_clean[['ApprovedBudgetForContract', 'ContractCost', 'CostVariance', 'CostVariancePct']].head(10))
```

```python
df_clean.to_csv('/content/drive/MyDrive/dpwh_flood_control_cleaned.csv', index=False)
```

```python
print(df_clean[['ApprovedBudgetForContract', 'ContractCost', 'CostVariance', 'CostVariancePct']].describe())
```

```python
region_counts = df_clean['Region'].value_counts()
print(region_counts)
```

```python
region_variance = df_clean.groupby('Region')['CostVariancePct'].mean()
region_variance = region_variance.sort_values(ascending=False)
print(region_variance)
```

```python
work_variance = df_clean.groupby('TypeOfWork')['CostVariancePct'].mean()
work_variance = work_variance.sort_values(ascending=False)
print(work_variance)
```

```python
yearly_variance = df_clean.groupby('FundingYear')['CostVariancePct'].mean()
yearly_variance
```

```python
outlier = df_clean[df_clean['CostVariancePct'] == df_clean['CostVariancePct'].min()]
print(outlier[['ProjectName', 'Region', 'ApprovedBudgetForContract', 'ContractCost', 'CostVariancePct']])
```

```python
over_budget = df_clean[df_clean['CostVariancePct'] > 0]
under_budget = df_clean[df_clean['CostVariancePct'] < 0]
exact_budget = df_clean[df_clean['CostVariancePct'] == 0]

print("Over budget:", len(over_budget))
print("Under budget:", len(under_budget))
print("Exact budget:", len(exact_budget))
```
